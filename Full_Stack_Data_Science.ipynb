```python
#probability modeling:
import scipy.stats as stats
bmi_pop_mean=dataset.bmi.mean()
bmi_pop_std=dataset.bmi.std()
z_score=(30-bmi_pop_mean)/bmi_pop_std

prob_less=stats.norm.cdf(z_score)
prob_less # 45.67% of the clients have a bmi less than or equal to 30
```

```python
z_score=(50-bmi_pop_mean)/bmi_pop_std
prob_greater=1-stats.norm.cdf(z_score)
num=stats.norm.ppf(0.95,loc=bmi_pop_mean,scale=bmi_pop_std)
num #95% of clients have a bmi of either less than or equal to 41 implying that greater than 42 might be an outlier
```
